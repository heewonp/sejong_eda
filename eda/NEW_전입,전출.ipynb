```python
import pathlib
import pandas as pd
import numpy as np

import json
```

```python
input_path = pathlib.Path('./data')
```

```python
전입자수 = pd.read_csv(input_path.joinpath('20.세종시_전입자수.csv'),encoding="utf-8")
전출자수 = pd.read_csv(input_path.joinpath('21.세종시_전출자수.csv'),encoding="utf-8")
```

```python
전입자수['년도']=0
전입자수=전입자수.astype({'년월':str})
전입자수=전입자수.astype({'년도':str})
전입자수['년도']=전입자수["년월"].str[0:4]

전출자수['년도']=0
전출자수=전출자수.astype({'년월':str})
전출자수=전출자수.astype({'년도':str})
전출자수['년도']=전출자수["년월"].str[0:4]
```

```python
#전입수는 비슷비슷
전입자수합계_년도별전입=전입자수.groupby(by=['년도']).sum().groupby(level=[0]).cumsum()
전입자수합계_년도별전입.sort_values('년도',ascending=True).head(20)
```

```python
#전출은 갈수록 증가하는중
전출자수합계_년도별전입=전출자수.groupby(by=['년도']).sum().groupby(level=[0]).cumsum()
전출자수합계_년도별전입.sort_values('년도',ascending=True).head(20)
```

<h1>세종시 내 이사분석</h1>

```python
전입_세종에서전출=전입자수[전입자수['전출시']=='세종특별자치시']
전출_세종으로전입=전출자수[전출자수['전입시']=='세종특별자치시']
```

```python
#세종에서 세종으로 전입/전출은 계속증가
전입_세종에서전출=전입_세종에서전출.groupby(by=['년도']).sum().groupby(level=[0]).cumsum()
전입_세종에서전출.sort_values('년도',ascending=True).head(20)
```

```python

전출_세종으로전입=전출_세종으로전입.groupby(by=['년도']).sum().groupby(level=[0]).cumsum()
전출_세종으로전입.sort_values('년도',ascending=True).head(20)
```

<h1>타지역 이사분석</h1>

```python
전입_타지역전출=전입자수[전입자수['전출시']!='세종특별자치시']#타지역->세종
전출_타지역전입=전출자수[전출자수['전입시']!='세종특별자치시']#세종->타지역
```

```python
전입_타지역전출
```

```python
#타지역에서 세종으로 전입하는 수는 증가하는듯 했으나 다시 감소
전입_타지역전출_년도별=전입_타지역전출.groupby(by=['년도']).sum().groupby(level=[0]).cumsum()
전입_타지역전출_년도별.sort_values('년도',ascending=True).head(20)
```

```python
#세종에서 타지역으로 전출하는 수는 갈수록 크게 증가중
전출_타지역전입_년도별=전출_타지역전입.groupby(by=['년도']).sum().groupby(level=[0]).cumsum()
전출_타지역전입_년도별.sort_values('년도',ascending=True)
```

```python
#타지역에서 세종으로 이사오는 곳은 대전, 충북, 충남이 압도적
전입_타지역전출_전출시=전입_타지역전출.groupby(by=['전출시']).sum().groupby(level=[0]).cumsum()
전입_타지역전출_전출시.sort_values('전입자수',ascending=False)
```
