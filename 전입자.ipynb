```python
#년도 변수 생성
전입자수['년도']=0
```

```python
#년도 인덱싱을 위해 str형으로 변환
전입자수=전입자수.astype({'년월':str})
전입자수.dtypes
```

```python
#년월에서 년도만 추출하여 년도변수에 추가
전입자수['년도']=전입자수["년월"].str[0:4]
```

```python

전입자수=전입자수.astype({'년월':str})
전입자수=전입자수.astype({'년도':str})
```

```python
#전입자수 합계 (조치원읍, 새롬동, 도담동)
전입자수합계_전입동=전입자수.groupby(by=['세종전입행정동']).sum().groupby(level=[0]).cumsum()
전입자수합계_전입동.sort_values('전입자수',ascending=False)
```

```python
#년도별 전입 2018-2017-2020 비슷비슷
전입자수합계_년도별전입=전입자수.groupby(by=['년도']).sum().groupby(level=[0]).cumsum()
전입자수합계_년도별전입.sort_values('전입자수',ascending=False).head(20)
```

```python
#세종 사이의 이사를 제외한 (대전25만, 충북23만, 충남22만 이사옴)
전입자수합계_전출시=전입자수.groupby(by=['전출시']).sum().groupby(level=[0]).cumsum()
전입자수합계_전출시.sort_values('전입자수',ascending=False)
```

```python
세종전출제외전입수=전입자수[전입자수['전출시']!='세종특별자치시']
```
